```python
from pygam import LinearGAM, s, f, l
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import LabelEncoder
```

## General additive model 


```python
data = pd.read_csv("Datasets/concrete.csv")
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

Składnia zgodna z sklearn. Model budujemy w stylu właściwym dla R, wskazując rozdzaj dopasowania dla kolejnych zmiennych:

s - spline
l - linear 
f - factor (kategorie)
te - tensor (wielowymiarowy)
fit_intercept - wyraz wolny (boolean)

```python
gam = LinearGAM(s(0,n_splines=10) + s(1,n_splines=10) + s(2,n_splines=5) + s(3,n_splines=5) + s(4,n_splines=5) + s(5,n_splines=4) + s(6,n_splines=10) + s(7,n_splines=10), fit_intercept=True)
gam = LinearGAM()
gam.fit(X, y)
gam.summary()
```

### Wykresy zależności cząstkowej (partial dependence)

Są to wykresy pozwalające śledzić wpływ (addytywny) danej zmiennej na wynik regresji

Jaki jest zakres wartości zmiennej zależnej? 0-80

```python
fig, axes = plt.subplots(2,4,figsize=(12,8))
for t, ax in zip(range(8),axes.flatten()):
    XX = gam.generate_X_grid(term=t)
    pdp, confi = gam.partial_dependence(term=t, X=XX, width=0.95)
    ax.plot(XX.T[t],pdp)
    ax.plot(XX.T[t],confi,c="r",ls=":")
    ax.set_title(X.columns[t])
```

### Tuning regesora
Przeszukiwanie losowej przestrzeni wartości lambda dla spline (stopnie dopasowania)
Dobierzemy przedział w zakresie -10, 10

```python
lams = np.random.rand(100,8) # random points on [0, 1], with shape (100, 8)
lams = lams * 4 - 2 # shift values to -10, 10
lams = 10 ** lams # transforms values to 1e-2, 1e2

gam = gam.gridsearch(X.values,y,lam=lams)
gam.summary()
```

```python
fig, axes = plt.subplots(2,4,figsize=(12,8))
for t, ax in zip(range(8),axes.flatten()):
    XX = gam.generate_X_grid(term=t)
    pdp, confi = gam.partial_dependence(term=t, X=XX, width=0.95)
    ax.plot(XX.T[t],pdp)
    ax.plot(XX.T[t],confi,c="r",ls=":")
    ax.set_title(X.columns[t])
    ax.set_ylim(-30,30)
```

### Przykład GAM dla danych zawierających kategorie (factor)

### Klasyfikacja dla wielu klas

Podobnie jak większość klasyfikatorów sklearn pygam nie obsługuje wieloklasowości bezpośrednio, a jedynie poprzez warper OneVsRestClassifier

```python
data = pd.read_csv("Datasets/iris.csv")
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
LE = LabelEncoder().fit(y)
y = LE.transform(y)

clf = OneVsRestClassifier(LogisticGAM()).fit(X, y)
```

```python
clf.estimators_ # mamy dostęp do poszczególnych estymatorów
```

Dzięki temu możemy pokazać funkcje zależności cząstkowej dla każdej klasy zmiennej zależnej i każdej zmiennej niezależnej.

```python
fig, axes = plt.subplots(3,4,figsize=(12,6))
for n,estimator in enumerate(clf.estimators_):
    axesrow=axes[n]
    for term, ax in enumerate(axesrow):
        XX = estimator.generate_X_grid(term=term)
        pdp, confi = estimator.partial_dependence(term=term, X=XX, width=0.95)
        ax.plot(XX.T[term],pdp)
        ax.plot(XX.T[term],confi,c="r",ls=":")
        ax.set_ylim(-20,20)
        
    axesrow[0].set_ylabel(LE.classes_[n])

for i,ax in enumerate(axes[0]):
    ax.set_title(X.columns[i])
```
